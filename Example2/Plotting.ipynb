```python
%matplotlib inline
from crlab_py.mpl import *
import numpy as np
import pandas as pd
import sqlalchemy as sa
import sqlite3
import scipy as sp
import scipy.signal
```

```python
conn = sqlite3.connect('sensors.db')
c = conn.cursor()
c.execute('SELECT value FROM tf_temp')
# print c.fetchall()

values = [float(x[0]) for x in c.fetchall()]
values = np.array(values)

# for testing, we only use the first 500 values
# values = values[0:500]
x = np.arange(0, values.size)
```

```python
# resample data to various numbers
print ('values', values.shape)
resampled_data = {}
for N in (10, 20, 50, 500, 1000):
    (dn, xn) = scipy.signal.resample(values, N, t=x)
    # dn = scipy.signal.decimate(x, 2, n=None, ftype='iir', axis=-1)
    # xn = np.linspace(0, values.size, dn.size)
    resampled_data[N] = (xn, dn)
    
fig, axes = plt.subplots(1 + len(resampled_data.keys()), figsize=(7, 6))
# plot original
ax = axes[0]
ax.plot(x, values)
ax.set_xlim(x.min(), x.max())

for ax, key in zip(axes[1:].flatten(), sorted(resampled_data.keys())):
    xn = resampled_data[key][0]
    dn = resampled_data[key][1]
    ax.plot(xn, dn, '.-')
    ax.set_xlim(xn.min(), xn.max())
    ax.set_title('N = {0}'.format(key))

fig.tight_layout()
```

```python
key = 500
xn = resampled_data[key][0]
dn = resampled_data[key][1]

f = scipy.interpolate.interp1d(xn, dn, kind='cubic')
# xnew = np.linspace(0, values.size, 10)
xnew = xn
values_splined = f(xnew)

fig, axes = plt.subplots(2, 1, figsize=(7, 3))
ax = axes[0]
ax.plot(xn, dn)
ax = axes[1]
ax.plot(xnew, values_splined)
```

```python
# Use a spline interpolation to get a smooth curve
from scipy.interpolate import UnivariateSpline

key = 50
xn = resampled_data[key][0]
dn = resampled_data[key][1]

spl = UnivariateSpline(xn, dn)
spl.set_smoothing_factor(0.5)
xs = np.linspace(0, values.size, 200)
values_splined = spl(xs)

fig, axes = plt.subplots(2, 1, figsize=(7, 3))
ax = axes[0]
ax.plot(x, values)

ax = axes[1]
ax.plot(xs, spl(xs))

for ax in axes:
    ax.set_xlim(0, x.max())
    ax.set_ylim(values.min(), values.max())
fig.tight_layout()
```
